# Nomalization



## 1.

for occupied house i, let 
$$ V_{i_{30}} = \{ \;\; j \; | \; j \in V \mbox{ and } d_{ij} <= 30 \} $$

and 

$$ V_{i_{60}} = \{ \;\; j \; | \; j \in V \mbox{ and } d_{ij} <= 60 \} $$

Our goal is let 

$$ \sum_{j \in V_{i_{30}} } w_{ij} = \alpha \times \sum_{j \in V_{i_{60}} } w_{ij}$$

such that we can compare these two objective functions. 

However, if $\;\; | V_{i_{60}} | >> | V_{i_{30}} | \;$ ,  which means we have many vacant houses around the occupied house i with distance = (30,60]. 

Therefore, $\alpha$ will dicrease and $\;\forall \;\;j \in V_{i_{30}}$ , $w_{ij} >> \alpha w_{ij} $ 

It turned out that the nomalized weight is controlled by the vacant houses set.


## 2. 

$$ \mbox{ if } x^2+y^2 \leq d^* \;\; , \;\;\; z = {1\over \sqrt{x^2 + y^2}} $$


$$ \hspace{-1.9cm}\mbox{ else } \hspace{3cm} z = 0 $$

Therefore

The volume is

$$ \int^{2\pi}_0  \int^{d^*}_0 {1\over t} t \; d \theta \; dt = \int^{2\pi}_0 d^* \; d\theta = 2\pi d^*$$


$$ \int^{d^*}_0 {1\over t} \times 2\pi t  \; dt = \int^{d^*}_0 2\pi dt = 2\pi d^*$$

Then we can always set a constant variable $\alpha = {1\over 2\pi d^*} $, let the totall valume alway equal to 1

```python
def getDistanceFromLatLonInKm(lat1,lon1,lat2,lon2):
    R = 6371 # Radius of the earth in km
    dLat = deg2rad(lat2-lat1)  ## deg2rad below
    dLon = deg2rad(lon2-lon1) 
    a = ( Math.sin(dLat/2) * Math.sin(dLat/2) +
        Math.cos(deg2rad(lat1)) * Math.cos(deg2rad(lat2)) * 
        Math.sin(dLon/2) * Math.sin(dLon/2) 
        ) 
    c = 2 * Math.atan2(Math.sqrt(a), Math.sqrt(1-a))
    d = R * c ## Distance in km
    return d

def deg2rad(deg):
    return deg * (Math.pi/180)
```
